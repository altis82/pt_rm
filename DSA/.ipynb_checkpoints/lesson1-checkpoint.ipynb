# Divide and conquer


## Merge sort
Divide the list into 2 parts, then continue dividing until 1 element. Merge 

```python
a =[2,3,5,1,6,7]
def merge_sort(list):
    if list is None:
        return []
    if len(list)<2:
        return list
    elif len(list)==2:
        if list[0]>list[1]:
            list[0], list[1]=list[1], list[0]
        return list
    else:
        
        mid =len(list)//2
        #print(f"mid {mid}")
        
        left_list=merge_sort(list[:mid])
        right_list= merge_sort(list[mid:])
        if left_list is None:
            left_list=[]
        if right_list is None:
            right_list=[]
        
        return merge(left_list, right_list)
        

def merge(l,r):
    tmp=[]
    i=j=0
    #print(f"left{l} right {r}")
    while i<len(l) and j<len(r):        
        if l[i]<=r[j]:
            tmp.append(l[i])
            i+=1
        else:
            tmp.append(r[j])
            j+=1
    tmp.extend(l[i:])
    tmp.extend(r[j:])
    return tmp
    
print(merge_sort(a))
    
```

## Quick sort
How to do a quick sort?
Quick sort is a divide and conquer alg by using a pivot element and partition the array into 2 parts
- left if less than pivot
- right if greater than pivot


```python

def quick_sort(a):
    if len(a)<=1:
        return a
    if len(a)==2:
        if a[0]>a[1]:
            a[0], a[1]=a[1], a[0]
    if len(a)>=3:
        pivot=len(a)//2
        
        left =[i for i in a if i<a[pivot]]
        mid =[i for i in a if i==a[pivot]]
        right =[i for i in a if i>a[pivot]]
        return quick_sort(left) + mid + quick_sort(right)

a=[2,3,1,4,3,5,2,4]
quick_sort(a)
a
```
